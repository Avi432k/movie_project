```python
# import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib 
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12, 8) # Adjusts the configuration of the plots we will make
```

```python
# Read in the data

df = pd.read_csv('/Users/avishandilya/Desktop/Data Analytics Project/ATAproject/Movie python project/movies.csv')
```

```python
# Looking at the data

df.head()
```

```python
# Getting basic info of data

df.info()
```

```python
# Check for missing values

df.isnull().sum()

for col in df.columns:
    pct_missing = (df[col].isnull().sum()/df[col].count())*100
    print('{} - {}%'.format(col, pct_missing)) 

```

```python
# Remove missing values

df.dropna(inplace=True)
```

```python
# Check for duplicates

df.duplicated().sum()
```

```python
# Data types for columns

df.dtypes
```

```python
# Change data type of columns

df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')

df
```

```python
# Change the format of the released column

df[['release_date', 'release_country']] = df['released'].str.split('(', expand=True)
df['release_country'] = df['release_country'].str.strip(')')
df['release_date'] = df['release_date'].str.strip()
df['release_date'] = pd.to_datetime(df['release_date'], format = '%B %d, %Y', errors = 'coerce')
#df['release_date'] = df['release_date'].dt.strftime('%Y-%m-%d')
#df = df.drop(columns=['release_date', 'release_country'])
df
```

```python
# Create column for correct year released based on the year shown in the released column

df['yearcorrected'] = df['release_date'].astype(str).str[:4]
df
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Scatter plot with budget vs gross 

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Film Budget')
plt.ylabel('Gross Earnings')
plt.show()
```

```python
# Plot budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={'color': 'blue'}, line_kws={'color': 'red'})
plt.show()
```

```python
# Looking at correlations

df.corr(numeric_only=True)
```

```python
corr_matrix = df.corr(numeric_only=True)

sns.heatmap(corr_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
# Changing object columns to numerical columns

df_numerised = df

for col in df_numerised.columns:
    if df_numerised[col].dtype == 'object':
        df_numerised[col] = df_numerised[col].astype('category')
        df_numerised[col] = df_numerised[col].cat.codes

df_numerised
```

```python
corr_matrix = df_numerised.corr(numeric_only=True)

sns.heatmap(corr_matrix, annot=True)
plt.title('Correlation Matrix for Movie Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
corr_mat = df_numerised.corr(numeric_only=True)
corr_pairs = corr_mat.unstack()
corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr_pairs = sorted_pairs[(sorted_pairs > 0.5) & (sorted_pairs < 0.98)]
high_corr_pairs
```
